```python
# Import libraries
import pandas as pd
from datetime import datetime
```

```python
# Load dataframe from csv
df = pd.read_csv("triangle-Table 1.csv")
df = df.set_index("Relating to Period")
df
```

```python
# A funcion that takes a date and quarter string (e.g. "2010 Q1") and a number of years (e.g. 3) and outputs a date string which can be used to index the table
def quarter_to_date(quarter_string, y):
    year, q = quarter_string.split()
    year = int(year)

    if q == 'Q1':
        dt = datetime(year+y, 5, 1)
    elif q == 'Q2':
        dt = datetime(year+y, 8, 1)
    elif q == 'Q3':
        dt = datetime(year+y, 11, 1)
    elif q == 'Q4':
        dt = datetime(year+y+1, 2, 1)
    else:
        raise ValueError(f"Invalid quarter format: {q}")
    
    return dt.strftime('%b-%y')

# This should output the date that corresponds to 3 years after 2010 Q1
date_result = quarter_to_date("2010 Q1", 3)
print(date_result)
```

```python
# Create a new dataframe that uses the three revisions as columns and each quarter as a row

# Define columns
columns = ["1_year_revision", "2_year_revision", "3_year_revision"]

# Create empty DataFrame
df_new = pd.DataFrame(columns=columns)

# Loop through each quarter (q)
for q in df.columns:

    # Get the index for this quarter after 0 years
    i0 = quarter_to_date(q, 0)
    # Index the dataframe to find the value at this quarter at the 0th year
    y_0 = float(df[q][i0])

    # Get the index for this quarter after 1 years
    i1 = quarter_to_date(q, 1)
    try:
        # Get the value for this quarter after 1 year
        y_1 = float(df[q][i1])
        # Calculate the difference between year 1 and year 0
        d1 = round(y_1-y_0, 1)
    except:
        d1 = None

    # Get the index for this quarter after 2 years
    i2 = quarter_to_date(q, 2)
    try:
        # Get the value for this quarter after 2 year
        y_2 = float(df[q][i2])
        # Calculate the difference between year 2 and year 0
        d2 = round(y_2-y_0, 1)
    except:
        d2 = None

    # Get the index for this quarter after 3 years
    i3 = quarter_to_date(q, 3)
    try:
        # Get the value for this quarter after 3 year
        y_3 = float(df[q][i3])
        # Calculate the difference between year 3 and year 0
        d3 = round(y_3-y_0, 1)
    except:
        d3 = None

    # Add a row to the new dataframe with these three differences
    df_new.loc[q] = [d1, d2, d3]
```

```python
# Add a new column 'average_revision' containing the average of the other three columns
df_new['average_revision'] = df_new[['1_year_revision', '2_year_revision', '3_year_revision']].mean(axis=1)
```
